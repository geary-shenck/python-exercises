```python
#Prompt the user to enter an integer.
#Display a table of squares and cubes from 1 to the value entered.
#Ask if the user wants to continue.
#Assume that the user will enter valid data.
#Only continue if the user agrees to.

import tabulate as tabbed

list1 =[[],[],[]]
user_want = False
while user_want != True:
    user_number5 = int(float(input("What whole number would you like to go up to?")))

    list1=([ele for ele in range(1,user_number5+1)],\
        [ele*ele for ele in range(1,user_number5+1)],\
        [ele*ele*ele for ele in range(1,user_number5+1)])
    print(list1)
    print("\n Here is your table\n")
    print('\n'\
        .join(['\t'.join([str(cell) for cell in row]) for row in list1]))
    print('\n requires user input')
    user_says = (input("Do you want to continue out of loop or 'no' to stay in loop? Yes/No")).upper()
    if user_says == "YES":
        user_want = True
```

```python
#Prompt the user for a numerical grade from 0 to 100.
#Display the corresponding letter grade.
#Prompt the user to continue.
#Assume that the user will enter valid integers for the grades.
#The application should only continue if the user agrees to.
#Grade Ranges:
#A : 100 - 88
#B : 87 - 80
#C : 79 - 67
#D : 66 - 60
#F : 59 - 0
cont = True
while cont == True:
    while True:
        try:
            num_grade = int(input("Numerical Grade?"))
        except ValueError:
            print ("try again")
            continue
        else:
            #successful type
            num_grade = int(float(num_grade))
            if num_grade <= 59:
                let_grade = "F"
            elif num_grade <= 62:
                let_grade = "D-"
            elif num_grade <= 66:
                let_grade = "D+"
            elif num_grade <= 70:
                let_grade = "C-"
            elif num_grade <= 79:
                let_grade = "C+"
            elif num_grade <= 84:
                let_grade = "B-"
            elif num_grade <= 87:
                let_grade = "B+"
            elif num_grade <= 93:
                let_grade = "A-"
            elif num_grade <= 100:
                let_grade = "A+"
            print(num_grade, " ", let_grade)
            break
    cont = bool(input("Continue? y/n") == "y")
```
